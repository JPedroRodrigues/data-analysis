1. (5,0 pontos) Usando o método scipy.stats.poisson da biblioteca SciPy, escreva
um programa para plotar os valores de uma distribuição Poisson com média 0.1.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss

plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (6, 4)

poisson = ss.poisson(mu = 0.1)
poisson_sample = poisson.rvs(size = 1000, random_state = 143)

end = np.max(poisson_sample) + 1
t = np.arange(0, end)

plt.hist(
        poisson_sample, bins = end, range = (0, end),
        density = True, align = 'left', rwidth = 0.75,
        label = "Sample", alpha = 0.75, color = 'Orange'
)

plt.scatter(
    t, poisson.pmf(t),
    label = 'PMF',
    color = 'Green'
)

plt.legend(loc = 'best')
plt.show()
```

2. (5,0 pontos) Usando o método scipy.stats.expon da biblioteca SciPy, escreva um programa
para plotar os valores de uma distribuição exponencial com média 2.

```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt


mean = 2
x = np.linspace(expon.ppf(0.001, scale=mean), expon.ppf(0.999, scale=mean), 1000)
y = expon.pdf(x, scale = mean)


data = expon.rvs(scale=mean, size=1000, random_state = 2)

plt.plot(x, y, 'red', lw=2.5, alpha=1, label='Exponencial com média 2')
plt.hist(data, density=True, bins=30, alpha=0.5, label='Histograma', color = 'darkblue')
plt.legend(loc='best')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição Exponencial')
plt.show()
```
