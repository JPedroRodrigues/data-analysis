O dataset planets.csv do diretório de datasets do seaborn
([https://github.com/mwaskom/seaborn-data](https://github.com/mwaskom/seaborn-data))
contém uma conjunto de dados sobre exoplanetas e os métodos usados em sua descoberta.

Um de seus atributos é o ano em que eles foram descobertos.

Carregue o dataset usando o comando load_dataset do seaborn.

Usando seaborn, crie um gráfico de série temporal com o formato de gáfico de barras, mostrando o ano em que eles foram descobertos.

```python
import seaborn as sns
import matplotlib as plt

sns.set_theme("paper", style="ticks")

planets = sns.load_dataset("planets")

x_ticks = [1989, 1997, 2002, 2007, 2012]
y_ticks = [0, 50, 100, 150, 200]

plot = sns.displot(
    data = planets, kind = "hist",
    x = "year", aspect=2,
    color="#05696b", shrink = .9
)


plot.set_axis_labels("Anos", "Quantidade de Planetas")
plot.set(
    xticks = x_ticks,
    yticks = y_ticks,
    title = "Total de Exoplanetas Descobertos ao longo dos Anos"
)
```
