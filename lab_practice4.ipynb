## How to Create a Probability Density Function Plot in Python with the Numpy, Scipy, and Matplotlib Modules
---

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-4, 4, 0.001)
plt.plot(x, norm.pdf(x))
```

## Plotting Distributions with matplotlib and scipy

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss


def plot_normal (x, mu = 0, sigma = 1, cdf = False, **kwargs):
  if cdf:
    y = ss.norm.cdf(x, mu, sigma)
  else:
    y = ss.norm.pdf(x, mu, sigma)

  plt.plot(x, y, **kwargs)


plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (12, 8)

x = np.linspace(-5, 5, 5000)

plot_normal(x)
plot_normal(x, cdf=True)
```
