# Aprendendo Seaborn com Seaborn

## Introdução ao Seaborn
---
Assim como Matplotlib, Seaborn é uma biblioteca utilizada para a criação de gráficos visuais em Python. Ele constroi sobre matplotlib e se integra fortemente com estruturas de dados em pandas.

Vamos, agora, a um exemplo do que o Seaborn é capaz de fazer.

```python
# Importa Seaborn
import seaborn as sns

# Aplica o tema padrão
sns.set_theme()

# Carregando um dataset predefinido
tips = sns.load_dataset("tips")

# Cria um gráfico de dispersão
sns.relplot(
    data = tips,
    x = "total_bill", y = "tip",
    col = "time", hue ="smoker",
    style = "smoker", size = "size"
)
```

Na prática, seaborn se utiliza de matplotlib para criar seus gráficos. Para gerar os gráficos de maneira interativa, utilize Jupyter ou Google Colab. Caso contrário, utilize o `matplotlib.pyplot.show()`.

### Temas

O comando `sns.set_theme()` possui diversos parâmetros úteis, que são:
- `darkgrid` (plano de fundo escuro com reticulado branco)
- `whitegrid` (plano de fundo branco com reticulado preto)
- `dark` (plano de fundo escuro e sem reticulado)
- `white` (plano de fundo claro)
- `ticks` (marcações nos dados dos eixos x e y)

### Removendo eixos

O comando `sns.despine()`. retira os eixos superior e esquerdo.

### Datasets

`sns.load_dataset()` acessa facilmente um conjunto de dados pré-definidos. Estes datases são dataframes do pandas, que poderiam facilmente ser lidos com `pandas.read_csv()`.

### Gráfico de Dispersão

O gráfico de dispersão foi facilmente criado com o comando `sns.relplot()`.

## API de alto nível para gráficos estatísticos
---

Não há um modo universalmente bom para representar distribuições, como também as relações entre dois dados. É por esta razão que `relplot()` possui "rel" (de relation) em seu nome. Podemos especificar o tipo de relação desejada por meio do parâmetro `kind`, como veremos a seguir:

```python
import seaborn as sns

dots = sns.load_dataset("dots")

sns.set_theme()

sns.relplot(
    data = dots, kind = "line",
    x = "time", y = "firing_rate",
    col = "align", hue = "choice",
    size = "coherence", style = "choice",
    facet_kws = dict(sharex=False)
)
```

## Estimativa Estatística

Frequentemente, estamos interessados no valor médio de uma variável como função para outra variável. O seaborn também nos ajuda nesta tarefa:

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")

sns.set_theme()

sns.relplot(
    data = fmri, kind = "line",
    x = "timepoint", y = "signal",
    col = "region", hue = "event",
    style = "event"
)
```

O seaborn é totalmente capaz de desenhar margens de erro automaticamente quando dados estatísticos são estimados.

Além disto, a estimativa estatística vai além da descrição. A seaborn consegue criar um modelo de regressão linear para gráficos de dispersão por meio da função `sns.lmplot()`:

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.lmplot(
    data = tips, x = "total_bill", y = "tip",
    col = "time", hue = "smoker"
)
```

## Representando Distribuições

Análise estatística requer conhecimento sobre a distribuição das variáveis de um conjunto de dados. A função `sns.displot()` suporta uma série de modos de representação de distribuições, como histogramas e gráficos de densidade.

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.set_theme("talk")

sns.displot(
    data= tips, x = "total_bill",
    col = "time", kde = True
)

sns.displot(
    data = tips, x = "total_bill",
    kind = "ecdf", col = "time",
    hue = "smoker", rug = True
)
```

## Gráficos para dados categóricos

`sns.catplot()` permite visualizar as distribuições geradas com variáveis categóricas de diversos modos:

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.set_style("whitegrid")


sns.catplot(
    data = tips, kind = "swarm",
    x = "day", y = "total_bill",
    hue = "smoker"
)
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.set_style("darkgrid")

sns.catplot(
    data = tips, kind = "violin",
    x = "day", y = "total_bill",
    hue = "smoker", split = True
)
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.set_theme()

sns.catplot(
    data = tips, kind = "bar",
    x = "day", y = "total_bill",
    hue = "smoker"
)
```
